```python
import os
import numpy as np
import pickle
import GetData
import netCDF4 as nc
import datetime
import time
```

```python
variables = []
variables_list = os.listdir("F:/MODIS-Aqua/")
for var in variables_list:
    file = "F:/MODIS-Aqua/"+var+"/"+os.listdir("F:/MODIS-Aqua/"+var)[0]
    with nc.Dataset(file,'r') as f:
        print(f.variables.keys())
        variables.append(list(f.variables.keys())[0])
print("\n\n",variables)
```

```python
with open("../0 save data/3 variables.pkl", 'wb') as f:
    pickle.dump(variables,f)
```

```python
use_data = "../0 save data/2 sorted_pigments.pkl"
pkl_path = "../0 save data/satellite grid data/"

if not os.path.exists(pkl_path): os.mkdir(pkl_path)
    
window_s = 3
window_t = 1
```

```python
root_path = "F:/MODIS-Aqua/"
file_list = os.listdir(root_path)
```

```python
for i,var in enumerate(file_list):
    if variables[i] == "sst": continue
    if variables[i] == "par": continue
    if variables[i] == "chlor_a": continue
    if variables[i] == "Zeu_lee": continue
    if variables[i] == "bbp_s_giop": continue
        
    var_path = root_path+var+"/"
    param = {"window_s":window_s,
         "window_t":window_t,
         "path":var_path,
         "use_data": use_data,
         "var":variables[i],
         "lon_var":"lon",
         "lat_var":"lat"
        }
    data = GetData.main(param)
    with open(pkl_path+var+".pkl", 'wb') as f:
        pickle.dump(data, f)
```
