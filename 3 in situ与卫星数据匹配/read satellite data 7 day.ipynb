```python
import os
import numpy as np
import pickle
import GetData
import netCDF4 as nc
import datetime
import time
```

```python
with open("../0 save data/3 variables.pkl", 'rb') as f:
    variables = pickle.load(f)
```

```python
use_data = "../0 save data/2 sorted_pigments.pkl"
pkl_path = "../0 save data/satellite grid data 7 day/"

if not os.path.exists(pkl_path): os.mkdir(pkl_path)
    
window_s = 3
window_t = 7
```

```python
root_path = "F:/MODIS-Aqua/"
file_list = os.listdir(root_path)
```

```python
# a = ["Rrs_412","Rrs_443","Rrs_469","Rrs_488","Rrs_531","Rrs_547","Rrs_555","Rrs_645","Rrs_667","Rrs_678"]
```

```python
for i,var in enumerate(file_list):
    if os.path.exists(pkl_path+var+"&"+variables[i]+".pkl"):
        continue
    else:
        print(var+"&"+variables[i],end="\t\t")
        
    var_path = root_path+var+"/"
    param = {"window_s":window_s,
         "window_t":window_t,
         "path":var_path,
         "use_data": use_data,
         "var":variables[i],
         "lon_var":"lon",
         "lat_var":"lat"
        }
    data = GetData.main(param)
    with open(pkl_path+var+"&"+variables[i]+".pkl", 'wb') as f:
        pickle.dump(data, f)
```
